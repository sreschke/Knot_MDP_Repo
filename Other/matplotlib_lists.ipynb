# An Ipython notebook to view matplotlib graphs of data collected in the Knot_MDP.py script

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
job_name="SliceEnv_try_0"
losses_file_name=job_name+"_losses"
lr_file_name=job_name+"_learning_rates"
eps_file_name=job_name+"_epsilons"
alphas_file_name=job_name+"_alphas"
betas_file_name=job_name+"_betas"


losses=np.load(losses_file_name)
learning_rates=np.load(lr_file_name)
epsilons=np.load(eps_file_name)
alphas=np.load(alphas_file_name)
betas=np.load(betas_file_name)
```

```python
x=np.arange(len(losses))
fig=plt.figure()
ax1 = plt.subplot(511)
ax2 = plt.subplot(512, sharex = ax1)
ax3 = plt.subplot(513, sharex = ax1)
ax4 = plt.subplot(514, sharex = ax1)
ax5 = plt.subplot(515, sharex = ax1)
plt.setp(ax1.get_xticklabels(), visible=False)
plt.setp(ax2.get_xticklabels(), visible=False)
ax1.plot(x, learning_rates, "r", label='Learning Rates')
ax1.legend(loc="upper right")
ax2.plot(x, epsilons, "b", label='Epsilons')
ax2.legend(loc="upper right")
ax3.semilogy(x, losses, "g", label="Losses")
ax3.legend(loc="lower left")
ax4.plot(x, alphas, "r", label='alphas')
ax4.legend(loc="upper right")
ax5.plot(x, betas, "r", label='betas')
ax5.legend(loc="upper right")
ax5.set_xlabel('(per 10000 epochs)')
ax1.set_title(job_name+" Results")

plt.show()
```
